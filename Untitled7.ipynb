```python
# ======= Full OOF pipeline: LGB (TF-IDF) + CatBoost (text/cat) + OOF per-bin calibration =======
# Paste & run. Requires: train.csv at /mnt/data/train.csv
# Optional: DO_IMAGES=True to include image embeddings (heavy). Default: False.

import os, re, time, gc, json
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.model_selection import KFold
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack, csr_matrix, issparse
import lightgbm as lgb
from catboost import CatBoostRegressor, Pool
import joblib

# ----------------- CONFIG -----------------
TRAIN_CSV = "train.csv"
OUT_DIR = "artifacts"
os.makedirs(OUT_DIR, exist_ok=True)

NFOLD = 3             # set to 3 for debug, 5 for final
RND = 42
TF_MAX_FEATS = 15000  # safer default
LGB_ROUNDS = 1000
CAT_ITERS = 1500
DO_IMAGES = False     # set True to include images (slow & heavy)
IMG_BATCH = 256
# ------------------------------------------

assert os.path.exists(TRAIN_CSV), f"Train file not found at {TRAIN_CSV}"

# ----------------- helper functions -----------------
def smape(y_true, y_pred, eps=1e-9):
    num = np.abs(y_pred - y_true)
    denom = (np.abs(y_pred) + np.abs(y_true)) / 2.0 + eps
    return np.mean(num / denom) * 100.0

def parse_ipq_safe(text):
    res = {'pack_count': np.nan, 'qty': np.nan, 'unit': np.nan}
    if not isinstance(text, str) or text.strip()=='':
        return res
    txt = text.lower()
    m = re.search(r'(\d+)\s*[x×]\s*(\d+(?:[.,]\d+)?)\s*(kg|g|gm|ml|l|litre|liter|pcs|pc|piece|count)?', txt)
    if m:
        try:
            res['pack_count'] = float(m.group(1))
            res['qty'] = float(m.group(2).replace(',','.'))
            res['unit'] = (m.group(3) or np.nan)
            return res
        except: pass
    m = re.search(r'pack\s*of\s*(\d+)', txt)
    if m:
        try:
            res['pack_count'] = float(m.group(1))
        except: pass
    m = re.search(r'(\d+(?:[.,]\d+)?)\s*(kg|g|gm|ml|l|litre|liter|pcs|pc|piece|count)?', txt)
    if m:
        try:
            res['qty'] = float(m.group(1).replace(',','.'))
            res['unit'] = (m.group(2) or np.nan)
        except: pass
    return res

def normalize_unit_qty(qty, unit):
    if qty is None or (isinstance(qty, float) and np.isnan(qty)): return np.nan
    try: q = float(qty)
    except: return np.nan
    u = str(unit).lower() if isinstance(unit, str) else ''
    if u in ['kg']: return q*1000.0
    if u in ['g','gm']: return q
    if u in ['l','litre','liter','liters']: return q*1000.0
    if u in ['ml']: return q
    if u in ['pcs','pc','piece','count']: return q
    return q

def extract_brand_simple(text):
    t = str(text).lower()
    m = re.search(r'brand[:\- ]+\s*([a-z0-9&]+)', t)
    if m: return m.group(1)
    toks = re.split(r'[\|\-–—:,.]', t)  # split at common separators
    first = toks[0].strip().split()
    return first[0] if len(first)>0 else 'unknown'

# ----------------- load data & features -----------------
df = pd.read_csv(TRAIN_CSV)
n = len(df)
print("Loaded train:", df.shape)
df['catalog_content'] = df.get('catalog_content','').fillna('').astype(str)
df['image_link'] = df.get('image_link','').fillna('').astype(str)

# parse IPQ safely and numeric text stats
parsed = df['catalog_content'].apply(parse_ipq_safe).apply(pd.Series)
df = pd.concat([df, parsed], axis=1)
df['qty_norm'] = df.apply(lambda r: normalize_unit_qty(r['qty'], r['unit']), axis=1)
df['total_qty'] = df.apply(lambda r: (r['pack_count']*r['qty_norm']) if (not np.isnan(r.get('pack_count', np.nan)) and not np.isnan(r.get('qty_norm', np.nan))) else (r.get('qty_norm', np.nan) if not np.isnan(r.get('qty_norm', np.nan)) else (r.get('pack_count', np.nan) if not np.isnan(r.get('pack_count', np.nan)) else np.nan)), axis=1)
df['is_pack'] = ((~df['pack_count'].isna()) & (df['pack_count'] > 1)).astype(int)
df['content_len'] = df['catalog_content'].apply(len)
df['word_count'] = df['catalog_content'].apply(lambda x: len(str(x).split()))
df['num_digits'] = df['catalog_content'].apply(lambda x: sum(c.isdigit() for c in str(x)))
df['num_caps'] = df['catalog_content'].apply(lambda x: sum(1 for c in str(x) if c.isupper()))

# brand
df['brand'] = df['catalog_content'].apply(extract_brand_simple)
top_brands = df['brand'].value_counts().head(400).index.tolist()
df['brand_small'] = df['brand'].apply(lambda x: x if x in top_brands else 'other')

# numeric features
num_features = ['content_len','word_count','num_digits','num_caps','pack_count','qty','qty_norm','total_qty','is_pack']
num_features = [c for c in num_features if c in df.columns]
print("Numeric features:", num_features)

# target log (we will train on log1p)
y_raw = df['price'].fillna(0).values
y_log = np.log1p(y_raw)

# ----------------- TF-IDF vectorizer (for LGB model) -----------------
tf = TfidfVectorizer(max_features=TF_MAX_FEATS, ngram_range=(1,2), min_df=3)
X_text = tf.fit_transform(df['catalog_content'].astype(str))
joblib.dump(tf, os.path.join(OUT_DIR, 'tf_for_lgb.joblib'))
print("TF-IDF shape:", X_text.shape)

# ----------------- optional image embeddings (disabled by default) -----------------
if DO_IMAGES:
    # lightweight image embedding pipeline (ResNet50 via timm)
    import requests
    from PIL import Image
    import timm, torch, torchvision.transforms as T
    device = "cuda" if torch.cuda.is_available() else "cpu"
    IMG_MODEL_NAME = "resnet50"
    img_model = timm.create_model(IMG_MODEL_NAME, pretrained=True, num_classes=0).eval().to(device)
    transform = T.Compose([T.Resize((224,224)), T.CenterCrop(224), T.ToTensor(),
                           T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])])
    img_embs = np.zeros((n, img_model.num_features), dtype=np.float32)
    def download_img(url, timeout=6):
        try:
            r = requests.get(url, timeout=timeout)
            if r.status_code==200:
                return Image.open(io.BytesIO(r.content)).convert("RGB")
        except:
            return None
        return None
    # compute embeddings in batches
    for i in range(0, n, IMG_BATCH):
        j = min(n, i+IMG_BATCH)
        imgs = []
        idxs = []
        for idx in range(i, j):
            url = df['image_link'].iat[idx]
            if not isinstance(url, str) or url.strip()=="":
                imgs.append(None)
            else:
                try:
                    im = download_img(url)
                    imgs.append(im)
                except:
                    imgs.append(None)
        # process valid images
        batch_tensors = []
        valid_idx = []
        for k,im in enumerate(imgs):
            if im is None:
                continue
            try:
                t = transform(im)
                batch_tensors.append(t)
                valid_idx.append(k)
            except:
                pass
        if len(batch_tensors) == 0:
            # leave zeros for this batch
            continue
        with torch.no_grad():
            tensor = torch.stack(batch_tensors).to(device)
            feats = img_model(tensor).cpu().numpy()
        vpos = 0
        for k in range(len(imgs)):
            if k in valid_idx:
                img_embs[i+k,:] = feats[vpos]
                vpos += 1
    np.save(os.path.join(OUT_DIR, "img_embs.npy"), img_embs)
    print("Saved image embeddings shape:", img_embs.shape)
else:
    img_embs = None

# ----------------- Prepare dense numeric matrix (for both models) -----------------
X_num = df[num_features].fillna(-1).astype(np.float32).values
scaler_path = os.path.join(OUT_DIR, "num_scaler.joblib")
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_num_scaled = scaler.fit_transform(X_num)
joblib.dump(scaler, scaler_path)

# ----------------- Setup K-Fold OOF containers -----------------
kf = KFold(n_splits=NFOLD, shuffle=True, random_state=RND)
oof_lgb = np.zeros(n)
oof_cat = np.zeros(n)

models_lgb = []
models_cat = []

fold_i = 0
for tr_idx, val_idx in kf.split(df):
    fold_i += 1
    print(f"\n=== Fold {fold_i}/{NFOLD} ===")
    # LGB: TF-IDF sparse + numeric (sparse combination)
    X_tr_text = X_text[tr_idx]
    X_val_text = X_text[val_idx]
    X_tr_num_sp = csr_matrix(X_num_scaled[tr_idx])
    X_val_num_sp = csr_matrix(X_num_scaled[val_idx])
    X_tr = hstack([X_tr_text, X_tr_num_sp])
    X_val = hstack([X_val_text, X_val_num_sp])
    if issparse(X_tr) and not isinstance(X_tr, csr_matrix):
        X_tr = X_tr.tocsr()
        X_val = X_val.tocsr()
    y_tr = y_log[tr_idx]
    y_val = y_log[val_idx]

    dtrain = lgb.Dataset(X_tr, label=y_tr)
    dval = lgb.Dataset(X_val, label=y_val)

    lgb_params = {
        'objective': 'regression',
        'metric': 'rmse',
        'learning_rate': 0.03,
        'num_leaves': 128,
        'feature_fraction': 0.8,
        'bagging_fraction': 0.8,
        'bagging_freq': 5,
        'verbosity': -1,
        'seed': RND
    }
    model_lgb = lgb.train(lgb_params, dtrain, num_boost_round=LGB_ROUNDS,
                          valid_sets=[dtrain, dval],
                          callbacks=[lgb.early_stopping(stopping_rounds=80), lgb.log_evaluation(period=200)])
    models_lgb.append(model_lgb)
    pred_val_log = model_lgb.predict(X_val, num_iteration=model_lgb.best_iteration)
    oof_lgb[val_idx] = np.expm1(pred_val_log)   # store on price scale to fit calibration later

    # CatBoost: use numeric + brand_small (categorical) + text (text feature)
    tr_df = df.iloc[tr_idx].reset_index(drop=True)
    vl_df = df.iloc[val_idx].reset_index(drop=True)
    # Ensure columns exist
    for c in num_features:
        if c not in tr_df.columns:
            tr_df[c] = -1; vl_df[c] = -1
    cat_cols = ['brand_small']
    text_cols = ['catalog_content']

    # create Pools
    Xtr_pool = Pool(data=tr_df[num_features + cat_cols + text_cols],
                    label=np.log1p(tr_df['price'].fillna(0).values),
                    cat_features=cat_cols, text_features=text_cols)
    Xval_pool = Pool(data=vl_df[num_features + cat_cols + text_cols],
                     label=np.log1p(vl_df['price'].fillna(0).values),
                     cat_features=cat_cols, text_features=text_cols)

    model_cat = CatBoostRegressor(iterations=CAT_ITERS, learning_rate=0.03, depth=8,
                                 loss_function='RMSE', eval_metric='RMSE',
                                 random_seed=RND, early_stopping_rounds=80, verbose=200)
    model_cat.fit(Xtr_pool, eval_set=Xval_pool, use_best_model=True)
    models_cat.append(model_cat)
    pred_val_log_cat = model_cat.predict(Xval_pool)
    oof_cat[val_idx] = np.expm1(pred_val_log_cat)  # price scale

    # cleanup
    del X_tr, X_val, dtrain, dval, Xtr_pool, Xval_pool
    gc.collect()

# ----------------- compute OOF SMAPEs (models & ensemble) -----------------
print("\nOOF SMAPE LGB:", smape(y_raw, oof_lgb))
print("OOF SMAPE CatBoost:", smape(y_raw, oof_cat))
ensemble_oof = 0.5*oof_lgb + 0.5*oof_cat
print("OOF SMAPE simple mean ensemble:", smape(y_raw, ensemble_oof))

# ----------------- OOF-based per-bin calibration (fit multipliers on OOF) -----------------
df_oof = pd.DataFrame({'price': y_raw, 'lgb': oof_lgb, 'cat': oof_cat})
df_oof['ensemble'] = ensemble_oof
df_oof['logp'] = np.log1p(df_oof['price'])
df_oof['bin'] = pd.qcut(df_oof['logp'], 10, duplicates='drop')

bin_scalers = {}
for b, g in df_oof.groupby('bin'):
    best_s = 1.0
    best_score = smape(g['price'].values, g['ensemble'].values)
    # small search around 0.6..1.6
    for s in np.linspace(0.7, 1.3, 61):
        sc = smape(g['price'].values, g['ensemble'].values * s)
        if sc < best_score:
            best_score, best_s = sc, s
    bin_scalers[str(b)] = float(best_s)
    print(f"bin {b} -> scale {best_s:.3f} (smape {best_score:.3f})")

# apply calibration to ensemble OOF
def apply_bin_scaling(preds, prices, bin_scalers):
    logp = np.log1p(prices)
    bins = pd.qcut(logp, 10, duplicates='drop')
    scaled = preds.copy()
    for i,b in enumerate(bins):
        s = bin_scalers.get(str(b), 1.0)
        scaled[i] = scaled[i] * s
    return scaled

calibrated_ensemble_oof = apply_bin_scaling(df_oof['ensemble'].values.copy(), df_oof['price'].values, bin_scalers)
print("OOF SMAPE calibrated ensemble:", smape(df_oof['price'].values, calibrated_ensemble_oof))

# ----------------- Save artifacts -----------------
joblib.dump(models_lgb, os.path.join(OUT_DIR, "models_lgb_oof.joblib"))
joblib.dump(models_cat, os.path.join(OUT_DIR, "models_cat_oof.joblib"))
joblib.dump(bin_scalers, os.path.join(OUT_DIR, "bin_scalers.joblib"))
joblib.dump(num_features, os.path.join(OUT_DIR, "num_features.joblib"))
joblib.dump(top_brands, os.path.join(OUT_DIR, "top_brands.joblib"))
print("Saved artifacts to", OUT_DIR)

print("\nDone. Summary:")
print(" LGB OOF SMAPE:", smape(y_raw, oof_lgb))
print(" CatBoost OOF SMAPE:", smape(y_raw, oof_cat))
print(" Ensemble OOF SMAPE:", smape(y_raw, ensemble_oof))
print(" Calibrated Ensemble OOF SMAPE:", smape(y_raw, calibrated_ensemble_oof))
```

```python
import re, math, os, gc, joblib
import numpy as np, pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import mean_squared_error
from scipy.sparse import hstack, csr_matrix
import lightgbm as lgb

TRAIN = "train.csv"
assert os.path.exists(TRAIN), "train.csv missing at train.csv"

df = pd.read_csv(TRAIN)
df['catalog_content'] = df.get('catalog_content','').fillna('').astype(str)

# -------------- 1) improved brand extraction ----------------
def clean_prefixes(text):
    t = text.lower().strip()
    # remove common leading labels
    t = re.sub(r'^(item name:|product name:|product:|item:|brand:)\s*', '', t)
    return t

def extract_brand_improved(text):
    t = clean_prefixes(str(text))
    # If there's "Brand - X" or "Brand: X"
    m = re.search(r'brand[:\-\s]+([A-Za-z0-9&\.\-]+)', t)
    if m:
        return m.group(1).lower()
    # often title format: "<Brand> <Other words>"
    # pick first token if it's a capitalized token or contains letters
    toks = re.split(r'[\|\-–—:,\n]', t)
    first = toks[0].strip().split()
    if len(first)>0:
        maybe = first[0]
        # ignore words like 'item','pack','case'
        if maybe.lower() in {'item','pack','case','box','set','bundle','combo','product'} and len(first)>1:
            return first[1].lower()
        return maybe.lower()
    return 'unknown'

df['brand_clean'] = df['catalog_content'].apply(extract_brand_improved)

# -------------- 2) improved IPQ parse and unit normalization ----------------
UNIT_MAP = {
    'kg': 1000.0, 'kilogram':1000.0, 'kilograms':1000.0,
    'g': 1.0, 'gm':1.0, 'gram':1.0, 'grams':1.0,
    'l': 1000.0, 'litre':1000.0, 'liter':1000.0,
    'ml': 1.0,
    'oz': 28.3495, 'ounce':28.3495, 'ounces':28.3495,
    'lb': 453.592, 'lbs':453.592, 'pound':453.592, 'pounds':453.592,
    'fl oz': 29.5735
}

def parse_ipq_enhanced(text):
    txt = str(text).lower()
    res = {'pack_count': np.nan, 'qty': np.nan, 'unit': np.nan}
    if not txt or txt.strip()=='':
        return res

    # 1) look for patterns like "6 x 250 ml", "3x250ml"
    m = re.search(r'(\d+)\s*[x×]\s*(\d+(?:[.,]\d+)?)\s*(kg|g|gm|ml|l|litre|liter|oz|ounce|ounces|lb|lbs|pound|fl oz|can|bottle|pack|case|pcs|pc|piece)?', txt)
    if m:
        try:
            res['pack_count'] = float(m.group(1))
            res['qty'] = float(m.group(2).replace(',', '.'))
            res['unit'] = (m.group(3) or np.nan)
            return res
        except:
            pass
    # 2) "pack of 12" or "case of 24"
    m = re.search(r'(pack|case|box|pallet|bundle|set)[\s\w]*of\s*(\d+)', txt)
    if m:
        try:
            res['pack_count'] = float(m.group(2))
        except: pass
    # 3) find single quantity with unit, e.g. "250 ml", "12 oz"
    m = re.search(r'(\d+(?:[.,]\d+)?)\s*(kg|g|gm|ml|l|litre|liter|oz|ounce|ounces|lb|lbs|pound|pounds|fl oz|can|bottle|pcs|pc|piece)\b', txt)
    if m:
        try:
            res['qty'] = float(m.group(1).replace(',', '.'))
            res['unit'] = m.group(2)
            return res
        except: pass
    # 4) look for dozen or single counts e.g., "dozen", "12 pack"
    m = re.search(r'(\d+)\s*(pack|pcs|pieces|pieces\)|count|ct\b)', txt)
    if m:
        try:
            res['pack_count'] = float(m.group(1))
            res['unit'] = 'pcs'
        except: pass
    return res

def qty_to_canonical(qty, unit):
    if qty is None or (isinstance(qty,float) and math.isnan(qty)): return np.nan
    u = str(unit).strip().lower() if unit else ''
    try:
        q = float(qty)
    except:
        return np.nan
    # if unit is bottle/can/pack treat as 'count'
    if u in {'can','bottle','pack','case','pcs','pc','piece'}:
        return q  # count
    # if units in dictionary convert to grams (for mass) or ml (for volume) - both numeric
    # we'll treat both as numeric quantity (grams/ml), price_per_unit will be price per gram/ml
    if u in UNIT_MAP:
        return q * UNIT_MAP[u]
    # if unit is empty but qty is large, maybe it's count
    return q

# apply parsers
parsed = df['catalog_content'].apply(parse_ipq_enhanced).apply(pd.Series)
df = pd.concat([df, parsed], axis=1)
df['qty_norm'] = df.apply(lambda r: qty_to_canonical(r.get('qty', np.nan), r.get('unit', np.nan)), axis=1)
# total qty: pack_count * qty_norm when both available; else fallback
def compute_total_qty(row):
    pc = row.get('pack_count', np.nan)
    qn = row.get('qty_norm', np.nan)
    # if pack_count present and qty_norm present
    if not pd.isna(pc) and not pd.isna(qn):
        return float(pc) * float(qn)
    # if only qty_norm present, use that (single item)
    if not pd.isna(qn):
        return float(qn)
    # if only pack_count present, treat pack_count as count
    if not pd.isna(pc):
        return float(pc)
    return np.nan

df['total_qty'] = df.apply(compute_total_qty, axis=1)

# -------------- 3) create price_per_unit where possible ----------------
df['price_filled'] = df['price'].fillna(0).astype(float)
# price per canonical unit (grams/ml/count): price / total_qty  (only when total_qty > 0)
df['ppu'] = np.where(df['total_qty']>0, df['price_filled'] / df['total_qty'], np.nan)
# log targets
df['log_price'] = np.log1p(df['price_filled'])
df['log_ppu'] = np.log1p(df['ppu'].fillna(0))  # 0 where missing -> we'll treat separately

# -------------- 4) bulk flag classifier ----------------
# heuristic: presence of keywords indicating bulk or 'case/pallet/pack of' or if price per unit is very small compared to price
bulk_keywords = r'\b(case|pallet|pack of|pack of \d+|bulk|bundle|box of|carton|pallet of|crate|packaged|case\sof)\b'
df['bulk_hint'] = df['catalog_content'].str.contains(bulk_keywords, flags=re.I, regex=True).astype(int)
# also bulk when pack_count >= 2
df['bulk_hint'] = np.where(df['pack_count'].fillna(0) >= 2, 1, df['bulk_hint'])

# LOSSOF: create a bulk label for model training (if high total price and pack_count>1 or bulk_hint)
# crude rule: if pack_count>1 or 'pack of' present OR price > 1000*median price -> bulk
median_price = df['price_filled'].median()
df['bulk_label'] = ((df['pack_count'].fillna(0) >= 2) | (df['catalog_content'].str.contains(r'pack of|case of|pallet of', case=False, regex=True)) | (df['price_filled'] > 50*median_price)).astype(int)

# ----------------- quick split for diagnostics ------------------
train_df, val_df = train_test_split(df, test_size=0.15, random_state=42, stratify=pd.qcut(df['log_price'].fillna(0), 10, duplicates='drop'))

print("Sizes:", df.shape, train_df.shape, val_df.shape)

# ----------------- features for models ----------------------
text_col = 'catalog_content'
tf = TfidfVectorizer(max_features=10000, ngram_range=(1,2), min_df=3)
X_tr_text = tf.fit_transform(train_df[text_col].astype(str))
X_val_text = tf.transform(val_df[text_col].astype(str))

# numeric features
num_feats = ['content_len','word_count','num_digits','num_caps','pack_count','qty_norm','total_qty','bulk_hint']
num_feats = [c for c in num_feats if c in train_df.columns]
X_tr_num = csr_matrix(train_df[num_feats].fillna(-1).values)
X_val_num = csr_matrix(val_df[num_feats].fillna(-1).values)

# brand one-hot (top 200)
topb = train_df['brand_clean'].value_counts().head(200).index
train_df['brand_small'] = train_df['brand_clean'].apply(lambda x: x if x in topb else 'other')
val_df['brand_small'] = val_df['brand_clean'].apply(lambda x: x if x in topb else 'other')
brand_tr = pd.get_dummies(train_df['brand_small'], prefix='brand')
brand_val = pd.get_dummies(val_df['brand_small'], prefix='brand').reindex(columns=brand_tr.columns, fill_value=0)

X_tr = hstack([X_tr_text, X_tr_num, csr_matrix(brand_tr.values)])
X_val = hstack([X_val_text, X_val_num, csr_matrix(brand_val.values)])

# ----------------- target: predict ppunit (preferred) if available, else price ------------------
# We'll fit model to predict log_ppu where ppu exists, otherwise we predict log_price and treat differently at inference
# Build mask for rows with total_qty
tr_mask_ppu = ~train_df['ppu'].isna() & (train_df['total_qty']>0)
val_mask_ppu = ~val_df['ppu'].isna() & (val_df['total_qty']>0)

print("Train has ppu:", tr_mask_ppu.sum(), "val has ppu:", val_mask_ppu.sum())

# Train LGB to predict log_ppu for rows that have ppu
dtrain = lgb.Dataset(X_tr[tr_mask_ppu.values], label=np.log1p(train_df.loc[tr_mask_ppu, 'ppu'].values))
dval = lgb.Dataset(X_val[val_mask_ppu.values], label=np.log1p(val_df.loc[val_mask_ppu, 'ppu'].values))

params = {
    'objective':'regression','metric':'rmse','learning_rate':0.05,
    'num_leaves':64,'feature_fraction':0.8,'verbosity':-1,'seed':42
}
model_ppu = lgb.train(params, dtrain, num_boost_round=800, valid_sets=[dtrain,dval],
                     callbacks=[lgb.early_stopping(stopping_rounds=80), lgb.log_evaluation(period=100)])
# predict on val ppu rows and compute SMAPE on price after conversion
pred_log_ppu_val = model_ppu.predict(X_val[val_mask_ppu.values], num_iteration=model_ppu.best_iteration)
pred_ppu_val = np.expm1(pred_log_ppu_val)
# reconstruct predicted price = pred_ppu * total_qty
pred_price_from_ppu = pred_ppu_val * val_df.loc[val_mask_ppu, 'total_qty'].values

# For val rows without total_qty, fallback to direct price model (train small model)
dtrain_price = lgb.Dataset(X_tr[~tr_mask_ppu.values], label=np.log1p(train_df.loc[~tr_mask_ppu, 'price_filled'].values))
dval_price = lgb.Dataset(X_val[~val_mask_ppu.values], label=np.log1p(val_df.loc[~val_mask_ppu, 'price_filled'].values))
model_price = lgb.train(params, dtrain_price, num_boost_round=800, valid_sets=[dtrain_price,dval_price],
                        callbacks=[lgb.early_stopping(stopping_rounds=80), lgb.log_evaluation(period=100)])
pred_log_price_val = model_price.predict(X_val[~val_mask_ppu.values], num_iteration=model_price.best_iteration)
pred_price_val_fallback = np.expm1(pred_log_price_val)

# combine predictions for val
pred_val_combined = np.zeros(len(val_df))
# fill ppu rows
pred_val_combined[val_mask_ppu.values] = pred_price_from_ppu
# fill fallback rows
pred_val_combined[~val_mask_ppu.values] = pred_price_val_fallback

# evaluate SMAPE on val
def smape(y_true, y_pred, eps=1e-9):
    return np.mean(np.abs(y_pred - y_true) / ((np.abs(y_pred) + np.abs(y_true))/2 + eps)) * 100.0

val_smape = smape(val_df['price_filled'].values, pred_val_combined)
print("VAL SMAPE (ppu model + fallback price model):", val_smape)

# Quick compare to baseline naive LGB on log_price (train on everything)
dtrain_all = lgb.Dataset(X_tr, label=np.log1p(train_df['price_filled'].values))
dval_all = lgb.Dataset(X_val, label=np.log1p(val_df['price_filled'].values))
model_all = lgb.train(params, dtrain_all, num_boost_round=800, valid_sets=[dtrain_all,dval_all],
                      callbacks=[lgb.early_stopping(stopping_rounds=80), lgb.log_evaluation(period=100)])
pred_all_log = model_all.predict(X_val, num_iteration=model_all.best_iteration)
pred_all = np.expm1(pred_all_log)
print("VAL SMAPE (baseline LGB price):", smape(val_df['price_filled'].values, pred_all))

# Save models & vectorizer
joblib.dump(model_ppu, "artifacts/model_ppu.joblib")
joblib.dump(model_price, "artifacts/model_price.joblib")
joblib.dump(model_all, "artifacts/model_all.joblib")
joblib.dump(tf, "artifacts/tf_ipq_fix.joblib")
print("Saved models to artifacts")
```

```python
import re, math, pandas as pd, numpy as np
TRAIN = "train.csv"
df = pd.read_csv(TRAIN)
df['catalog_content'] = df.get('catalog_content','').fillna('').astype(str)

# unit detection helper
def detect_unit_type(text):
    t = text.lower()
    # mass units
    if re.search(r'\b(kg|g|gm|gram|grams|oz|ounce|pound|lb|lbs)\b', t):
        return 'mass'
    # volume units
    if re.search(r'\b(ml|l|liter|litre|fl oz|fluid ounce)\b', t):
        return 'volume'
    # count-like
    if re.search(r'\b(pack|case|pallet|box|bottle|can|pcs|pc|piece|dozen)\b', t):
        return 'count'
    return 'unknown'

df['unit_type'] = df['catalog_content'].apply(detect_unit_type)

# quick parse of pack_count and qty (the same parse you used)
def quick_ipq(text):
    m = re.search(r'(\d+)\s*[x×]\s*(\d+(?:[.,]\d+)?)', text.lower())
    pack = None; qty=None
    if m:
        pack = float(m.group(1))
        qty = float(m.group(2).replace(',', '.'))
    else:
        m2 = re.search(r'pack\s*of\s*(\d+)', text.lower())
        if m2:
            pack = float(m2.group(1))
    return pack, qty

parsed = df['catalog_content'].apply(lambda t: quick_ipq(str(t)))
df['pack_count'] = parsed.apply(lambda x: x[0])
df['qty_heur'] = parsed.apply(lambda x: x[1])

# compute total_qty_estimate (very rough): pack_count * qty_heur if both present, else qty_heur, else NaN
df['total_qty_est'] = df.apply(lambda r: (r['pack_count']*r['qty_heur']) if (not pd.isna(r['pack_count']) and not pd.isna(r['qty_heur'])) else (r['qty_heur'] if not pd.isna(r['qty_heur']) else np.nan), axis=1)

# price per estimated unit
df['price'] = df['price'].fillna(0).astype(float)
df['ppu_est'] = np.where(df['total_qty_est']>0, df['price'] / df['total_qty_est'], np.nan)

print("Unit type counts:\n", df['unit_type'].value_counts())
print("\nTotal_qty_est stats (non-null):")
print(df['total_qty_est'].describe())
print("\nPPU_est stats (non-null):")
print(df['ppu_est'].describe())

# show top problematic rows where total_qty_est is huge or tiny
print("\nTop rows with huge total_qty_est (>99.9 pct):")
print(df[df['total_qty_est'] > df['total_qty_est'].quantile(0.999)][['catalog_content','price','total_qty_est']].head(20))

print("\nTop rows with tiny nonzero total_qty_est (<0.001 pct quantile):")
print(df[(df['total_qty_est']>0) & (df['total_qty_est'] < df['total_qty_est'].quantile(0.001))][['catalog_content','price','total_qty_est']].head(20))
```

```python
# Cell 1 - config & imports
import os, re, math, json, time, gc
from pathlib import Path
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.model_selection import KFold, train_test_split
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import mean_squared_error
from scipy.sparse import hstack, csr_matrix, issparse
import joblib

# ML libs
import lightgbm as lgb
from catboost import CatBoostRegressor, Pool

# Embedding libs (optional)
from sentence_transformers import SentenceTransformer

# image libs used only if DO_IMAGES=True (timm + torch)
# import timm, torch, torchvision.transforms as T  # inside DO_IMAGES block

# ---------------- CONFIG ----------------
TRAIN_CSV = "train.csv"
OUT_DIR = "final_artifacts"
os.makedirs(OUT_DIR, exist_ok=True)

NFOLD = 3                # set 3 for debug, 5 for final
RND = 42

TF_MAX_FEATS = 15000     # reduce if OOM
SBERT_MODEL = "sentence-transformers/all-MiniLM-L6-v2"
SBERT_BATCH = 2048       # reduce if OOM
DO_IMAGES = False        # set True to enable image embeddings (slow)
IMG_MODEL = "resnet50"
IMG_BATCH = 256
DO_TFIDF_FOR_LGB = True  # TF-IDF + LGB baseline
SEED = RND

print("CONFIG: NFOLD", NFOLD, "TF_MAX_FEATS", TF_MAX_FEATS, "DO_IMAGES", DO_IMAGES)
```

```python
# Cell 2 - feature engineering
assert os.path.exists(TRAIN_CSV), f"Train file not found at {TRAIN_CSV}"
df = pd.read_csv(TRAIN_CSV)
n = len(df)
print("Loaded", df.shape)

# safe text fill
df['catalog_content'] = df.get('catalog_content','').fillna('').astype(str)
df['image_link'] = df.get('image_link','').fillna('').astype(str)
df['sample_id'] = df.get('sample_id', pd.Series(range(len(df))))

# ---- brand cleaning ----
def clean_prefixes(text):
    t = str(text).lower().strip()
    t = re.sub(r'^(item name:|product name:|product:|item:|brand:)\s*', '', t)
    return t

def extract_brand(text):
    t = clean_prefixes(str(text))
    m = re.search(r'brand[:\-\s]+([A-Za-z0-9&\.\-]+)', t)
    if m: return m.group(1).lower()
    toks = re.split(r'[\|\-–—:,\n]', t)
    first = toks[0].strip().split()
    if len(first)>0:
        maybe = first[0].lower()
        if maybe in {'item','pack','case','box','set','bundle','combo','product'} and len(first)>1:
            return first[1].lower()
        return maybe
    return 'unknown'

df['brand_clean'] = df['catalog_content'].apply(extract_brand)

# ---- enhanced IPQ parsing ----
UNIT_MAP = {
    'kg': 1000.0, 'kilogram':1000.0, 'kilograms':1000.0,
    'g': 1.0, 'gm':1.0, 'gram':1.0, 'grams':1.0,
    'l': 1000.0, 'litre':1000.0, 'liter':1000.0,
    'ml': 1.0,
    'oz': 28.3495, 'ounce':28.3495, 'ounces':28.3495,
    'lb': 453.592, 'lbs':453.592, 'pound':453.592, 'pounds':453.592,
    'fl oz': 29.5735
}

def parse_ipq(text):
    txt = str(text).lower()
    res = {'pack_count': np.nan, 'qty': np.nan, 'unit': np.nan}
    if not txt or txt.strip()=='':
        return res
    m = re.search(r'(\d+)\s*[x×]\s*(\d+(?:[.,]\d+)?)\s*(kg|g|gm|ml|l|litre|liter|oz|ounce|ounces|lb|lbs|pound|fl oz|can|bottle|pack|case|pcs|pc|piece)?', txt)
    if m:
        try:
            res['pack_count'] = float(m.group(1))
            res['qty'] = float(m.group(2).replace(',', '.'))
            res['unit'] = (m.group(3) or np.nan)
            return res
        except:
            pass
    m = re.search(r'(pack|case|box|pallet|bundle|set)[\s\w]*of\s*(\d+)', txt)
    if m:
        try:
            res['pack_count'] = float(m.group(2))
        except:
            pass
    m = re.search(r'(\d+(?:[.,]\d+)?)\s*(kg|g|gm|ml|l|litre|liter|oz|ounce|ounces|lb|lbs|pound|pounds|fl oz|can|bottle|pcs|pc|piece)\b', txt)
    if m:
        try:
            res['qty'] = float(m.group(1).replace(',', '.'))
            res['unit'] = m.group(2)
            return res
        except:
            pass
    m = re.search(r'(\d+)\s*(pack|pcs|pieces|count|ct)\b', txt)
    if m:
        try:
            res['pack_count'] = float(m.group(1)); res['unit']='pcs'
        except:
            pass
    return res

def qty_to_canonical(qty, unit):
    if qty is None or (isinstance(qty,float) and np.isnan(qty)): return np.nan
    try: q = float(qty)
    except: return np.nan
    u = str(unit).strip().lower() if unit else ''
    if u in {'can','bottle','pack','case','pcs','pc','piece','box','carton'}:
        return q  # count
    if u in UNIT_MAP:
        return q * UNIT_MAP[u]
    return q

parsed = df['catalog_content'].apply(parse_ipq).apply(pd.Series)
df = pd.concat([df, parsed], axis=1)
df['qty_norm'] = df.apply(lambda r: qty_to_canonical(r.get('qty', np.nan), r.get('unit', np.nan)), axis=1)

def compute_total_qty(row):
    pc = row.get('pack_count', np.nan)
    qn = row.get('qty_norm', np.nan)
    if not pd.isna(pc) and not pd.isna(qn):
        return float(pc) * float(qn)
    if not pd.isna(qn):
        return float(qn)
    if not pd.isna(pc):
        return float(pc)
    return np.nan

df['total_qty'] = df.apply(compute_total_qty, axis=1)
df['is_pack'] = ((~df['pack_count'].isna()) & (df['pack_count'] > 1)).astype(int)

# --- text stats ---
df['content_len'] = df['catalog_content'].apply(len)
df['word_count'] = df['catalog_content'].apply(lambda x: len(str(x).split()))
df['num_digits'] = df['catalog_content'].apply(lambda x: sum(c.isdigit() for c in str(x)))
df['num_caps'] = df['catalog_content'].apply(lambda x: sum(1 for c in str(x) if c.isupper()))

# ---- top brands list (for later) ----
top_brands = df['brand_clean'].value_counts().head(400).index.tolist()
df['brand_small'] = df['brand_clean'].apply(lambda x: x if x in top_brands else 'other')

# numeric feature list
num_features = ['content_len','word_count','num_digits','num_caps','pack_count','qty_norm','total_qty','is_pack']
num_features = [c for c in num_features if c in df.columns]

print("Numeric features:", num_features)
print("Sample rows:", df.head(2).to_dict(orient='records')[:2])

# save df for next cells
df.to_pickle(os.path.join(OUT_DIR, "df_fe.pkl"))
print("Saved FE df to", os.path.join(OUT_DIR, "df_fe.pkl"))
```

```python
# Cell 3 - SBERT embeddings
df = pd.read_pickle(os.path.join(OUT_DIR, "df_fe.pkl"))
n = len(df)
SBERT_OUT = os.path.join(OUT_DIR, "text_embeddings.npy")
print("SBERT model:", SBERT_MODEL)
sbert = SentenceTransformer(SBERT_MODEL)
sbert.max_seq_length = 256

if not os.path.exists(SBERT_OUT):
    emb_dim = sbert.get_sentence_embedding_dimension()
    emb = np.zeros((n, emb_dim), dtype=np.float32)
    for i in range(0, n, SBERT_BATCH):
        j = min(n, i+SBERT_BATCH)
        texts = df['catalog_content'].iloc[i:j].tolist()
        emb[i:j] = sbert.encode(texts, show_progress_bar=False, batch_size=SBERT_BATCH)
        print("encoded", i, j)
    np.save(SBERT_OUT, emb)
    print("Saved SBERT embeddings:", SBERT_OUT)
else:
    emb = np.load(SBERT_OUT)
    print("Loaded SBERT embeddings:", emb.shape)

print("Text embeddings shape:", emb.shape)
del sbert
gc.collect()
```

```python
# Cell 5 - prepare matrices for OOF
df = pd.read_pickle(os.path.join(OUT_DIR, "df_fe.pkl"))
y_raw = df['price'].fillna(0).values
y_log = np.log1p(y_raw)

# TF-IDF for LGB (sparse)
if DO_TFIDF_FOR_LGB:
    tf = TfidfVectorizer(max_features=TF_MAX_FEATS, ngram_range=(1,2), min_df=3)
    X_tfidf = tf.fit_transform(df['catalog_content'].astype(str))
    joblib.dump(tf, os.path.join(OUT_DIR, "tf_for_lgb.joblib"))
    print("TF-IDF shape", X_tfidf.shape)
else:
    X_tfidf = None

# numeric matrix
num_X = df[num_features].fillna(-1).astype(np.float32).values
scaler = StandardScaler()
num_X_scaled = scaler.fit_transform(num_X)
joblib.dump(scaler, os.path.join(OUT_DIR, "num_scaler.joblib"))

# text emb loaded from file
text_emb = np.load(os.path.join(OUT_DIR, "text_embeddings.npy"))


print("num_X shape", num_X_scaled.shape, "text_emb shape", text_emb.shape if text_emb is not None else None)
```

```python
# Cell 6 - NFOLD OOF training
kf = KFold(n_splits=NFOLD, shuffle=True, random_state=RND)

oof_lgb = np.zeros(n)
oof_cat = np.zeros(n)
models_lgb = []
models_cat = []

fold = 0
for tr_idx, val_idx in kf.split(df):
    fold += 1
    print(f"=== Fold {fold}/{NFOLD} ===")
    # LGB: TF-IDF sparse + numeric (sparse)
    if DO_TFIDF_FOR_LGB:
        X_tr_text = X_tfidf[tr_idx]; X_val_text = X_tfidf[val_idx]
        X_tr_num_sp = csr_matrix(num_X_scaled[tr_idx]); X_val_num_sp = csr_matrix(num_X_scaled[val_idx])
        X_tr = hstack([X_tr_text, X_tr_num_sp])
        X_val = hstack([X_val_text, X_val_num_sp])
        if issparse(X_tr) and not isinstance(X_tr, csr_matrix):
            X_tr = X_tr.tocsr(); X_val = X_val.tocsr()
    else:
        # alternative: use SBERT only as features for LGB
        X_tr = csr_matrix(np.hstack([text_emb[tr_idx], num_X_scaled[tr_idx]]))
        X_val = csr_matrix(np.hstack([text_emb[val_idx], num_X_scaled[val_idx]]))

    y_tr_log = y_log[tr_idx]; y_val_log = y_log[val_idx]

    dtrain = lgb.Dataset(X_tr, label=y_tr_log)
    dval = lgb.Dataset(X_val, label=y_val_log)

    lgb_params = {
        'objective':'regression','metric':'rmse','learning_rate':0.03,
        'num_leaves':128,'feature_fraction':0.8,'bagging_fraction':0.8,'bagging_freq':5,
        'verbosity':-1,'seed':RND
    }
    model_lgb = lgb.train(lgb_params, dtrain, num_boost_round=1000, valid_sets=[dval],
                          callbacks=[lgb.early_stopping(stopping_rounds=80), lgb.log_evaluation(period=200)])
    models_lgb.append(model_lgb)
    pred_val_log = model_lgb.predict(X_val, num_iteration=model_lgb.best_iteration)
    oof_lgb[val_idx] = np.expm1(pred_val_log)

    # CatBoost: use numeric + brand_small (cat) + SBERT embeddings concatenated to numeric
    tr_df = df.iloc[tr_idx].reset_index(drop=True); vl_df = df.iloc[val_idx].reset_index(drop=True)
    cat_cols = ['brand_small']
    text_cols = []  # we will pass SBERT embeddings directly rather than raw text for CatBoost (faster)
    # Build CatBoost numeric matrix = [num_scaled + text_emb]
    X_tr_cat = np.hstack([num_X_scaled[tr_idx], text_emb[tr_idx]])
    X_val_cat = np.hstack([num_X_scaled[val_idx], text_emb[val_idx]])
    # But CatBoost expects DataFrame with categorical columns included; we'll create DF for Pools
    cat_train_df = pd.DataFrame(X_tr_cat, columns=[f"f{i}" for i in range(X_tr_cat.shape[1])])
    cat_val_df = pd.DataFrame(X_val_cat, columns=[f"f{i}" for i in range(X_val_cat.shape[1])])
    # add brand col
    cat_train_df['brand_small'] = tr_df['brand_small'].astype(str).values
    cat_val_df['brand_small'] = vl_df['brand_small'].astype(str).values
    train_pool = Pool(data=cat_train_df, label=np.log1p(tr_df['price'].fillna(0).values), cat_features=['brand_small'])
    val_pool = Pool(data=cat_val_df, label=np.log1p(vl_df['price'].fillna(0).values), cat_features=['brand_small'])

    model_cat = CatBoostRegressor(iterations=1500, learning_rate=0.03, depth=8, loss_function='RMSE',
                                 eval_metric='RMSE', random_seed=RND, early_stopping_rounds=80, verbose=200)
    model_cat.fit(train_pool, eval_set=val_pool, use_best_model=True)
    models_cat.append(model_cat)
    pred_val_cat_log = model_cat.predict(val_pool)
    oof_cat[val_idx] = np.expm1(pred_val_cat_log)

    # cleanup per fold
    del X_tr, X_val, dtrain, dval, train_pool, val_pool
    gc.collect()

# Save OOFs and models
np.save(os.path.join(OUT_DIR, "oof_lgb.npy"), oof_lgb)
np.save(os.path.join(OUT_DIR, "oof_cat.npy"), oof_cat)
joblib.dump(models_lgb, os.path.join(OUT_DIR, "models_lgb.joblib"))
joblib.dump(models_cat, os.path.join(OUT_DIR, "models_cat.joblib"))
print("Saved OOFs & models to", OUT_DIR)

# diagnostics
def smape(y_true, y_pred, eps=1e-9):
    return np.mean(np.abs(y_pred - y_true) / ((np.abs(y_pred) + np.abs(y_true))/2 + eps)) * 100.0

print("OOF SMAPE LGB:", smape(y_raw, oof_lgb))
print("OOF SMAPE Cat:", smape(y_raw, oof_cat))
ensemble_oof = 0.5*oof_lgb + 0.5*oof_cat
print("OOF SMAPE ensemble mean:", smape(y_raw, ensemble_oof))
```

```python
# Cell 7 - stack & calibrate
oof_lgb = np.load(os.path.join(OUT_DIR, "oof_lgb.npy"))
oof_cat = np.load(os.path.join(OUT_DIR, "oof_cat.npy"))
X_meta = np.vstack([oof_lgb, oof_cat]).T
# simple ridge stacking (use cross-validated alphas)
alphas = [0.1, 1.0, 10.0]
stacker = RidgeCV(alphas=alphas, cv=5)
stacker.fit(X_meta, y_raw)  # meta-model predicts price scale directly
meta_oof = stacker.predict(X_meta)
print("Stack OOF SMAPE:", smape(y_raw, meta_oof))

# Per-bin calibration on meta_oof
df_oof = pd.DataFrame({'price': y_raw, 'meta': meta_oof})
df_oof['logp'] = np.log1p(df_oof['price'])
df_oof['bin'] = pd.qcut(df_oof['logp'], 10, duplicates='drop')

bin_scalers = {}
for b, g in df_oof.groupby('bin'):
    best_s, best_score = 1.0, smape(g['price'].values, g['meta'].values)
    for s in np.linspace(0.7, 1.3, 61):
        sc = smape(g['price'].values, g['meta'].values * s)
        if sc < best_score:
            best_score, best_s = sc, s
    bin_scalers[str(b)] = float(best_s)
    print("bin", b, "->", best_s, "smape", best_score)

# apply bin scaling
def apply_bin_scaling(preds, prices, bin_scalers):
    logp = np.log1p(prices)
    bins = pd.qcut(logp, 10, duplicates='drop')
    scaled = preds.copy()
    for i,b in enumerate(bins):
        s = bin_scalers.get(str(b), 1.0)
        scaled[i] = scaled[i] * s
    return scaled

meta_oof_cal = apply_bin_scaling(meta_oof.copy(), y_raw, bin_scalers)
print("Stacked + calibrated OOF SMAPE:", smape(y_raw, meta_oof_cal))

# Save stacker & scalers
joblib.dump(stacker, os.path.join(OUT_DIR, "stacker_ridge.joblib"))
joblib.dump(bin_scalers, os.path.join(OUT_DIR, "bin_scalers_meta.joblib"))
print("Saved stacker and bin scalers")
```

```python
# Cell 8 - inference (test)
TEST_CSV = "sample_test.csv"
if not os.path.exists(TEST_CSV):
    print("No test.csv found at test.csv — skip test prediction.")
else:
    test_df = pd.read_csv(TEST_CSV)
    test_df['catalog_content'] = test_df.get('catalog_content','').fillna('').astype(str)
    test_df['image_link'] = test_df.get('image_link','').fillna('').astype(str)
    # replicate FE steps for test
    test_df['brand_clean'] = test_df['catalog_content'].apply(extract_brand)
    parsed_test = test_df['catalog_content'].apply(parse_ipq).apply(pd.Series)
    test_df = pd.concat([test_df, parsed_test], axis=1)
    test_df['qty_norm'] = test_df.apply(lambda r: qty_to_canonical(r.get('qty', np.nan), r.get('unit', np.nan)), axis=1)
    test_df['total_qty'] = test_df.apply(compute_total_qty, axis=1)
    test_df['is_pack'] = ((~test_df['pack_count'].isna()) & (test_df['pack_count'] > 1)).astype(int)
    test_df['content_len'] = test_df['catalog_content'].apply(len)
    test_df['word_count'] = test_df['catalog_content'].apply(lambda x: len(str(x).split()))
    test_df['num_digits'] = test_df['catalog_content'].apply(lambda x: sum(c.isdigit() for c in str(x)))
    test_df['num_caps'] = test_df['catalog_content'].apply(lambda x: sum(1 for c in str(x) if c.isupper()))
    test_df['brand_small'] = test_df['brand_clean'].apply(lambda x: x if x in top_brands else 'other')

    # prepare features
    # TF-IDF
    if DO_TFIDF_FOR_LGB:
        tf = joblib.load(os.path.join(OUT_DIR, "tf_for_lgb.joblib"))
        X_test_text = tf.transform(test_df['catalog_content'].astype(str))
        X_test_num = csr_matrix(scaler.transform(test_df[num_features].fillna(-1).astype(float).values))
        X_test_lgb = hstack([X_test_text, X_test_num])
    else:
        X_test_lgb = csr_matrix(np.hstack([np.load(os.path.join(OUT_DIR,"text_embeddings.npy")), scaler.transform(test_df[num_features].fillna(-1).astype(float).values)]))

    # LGB predict: average predictions from models
    models_lgb = joblib.load(os.path.join(OUT_DIR, "models_lgb.joblib"))
    preds_lgb_log = np.zeros(len(test_df))
    for model in models_lgb:
        preds_lgb_log += model.predict(X_test_lgb, num_iteration=model.best_iteration) / len(models_lgb)
    preds_lgb = np.expm1(preds_lgb_log)

    # CatBoost predict
    models_cat = joblib.load(os.path.join(OUT_DIR, "models_cat.joblib"))
    test_num_scaled = scaler.transform(test_df[num_features].fillna(-1).astype(float).values)
    text_emb_test = np.load(os.path.join(OUT_DIR, "text_embeddings.npy"))[:len(test_df)]
    X_cat_arr = np.hstack([test_num_scaled, text_emb_test])
    cat_test_df = pd.DataFrame(X_cat_arr, columns=[f"f{i}" for i in range(X_cat_arr.shape[1])])
    cat_test_df['brand_small'] = test_df['brand_small'].astype(str).values
    preds_cat_log = np.zeros(len(test_df))
    for model in models_cat:
        preds_cat_log += model.predict(cat_test_df) / len(models_cat)
    preds_cat = np.expm1(preds_cat_log)

    # meta stack
    stacker = joblib.load(os.path.join(OUT_DIR, "stacker_ridge.joblib"))
    meta_features_test = np.vstack([preds_lgb, preds_cat]).T
    preds_meta = stacker.predict(meta_features_test)

    # apply bin scalers
    bin_scalers = joblib.load(os.path.join(OUT_DIR, "bin_scalers_meta.joblib"))
    prices_test = preds_meta.copy()
    # if test price unknown, we approximate logp by preds_meta when computing bins (practical)
    pred_calibrated = []
    logp = np.log1p(prices_test)
    bins = pd.qcut(logp, 10, duplicates='drop')
    for i,b in enumerate(bins):
        s = bin_scalers.get(str(b), 1.0)
        pred_calibrated.append(prices_test[i] * s)
    preds_final = np.maximum(np.array(pred_calibrated), 0.01)

    sub = pd.DataFrame({'sample_id': test_df['sample_id'], 'price': preds_final})
    sub.to_csv('test_out_final.csv', index=False)
    print("Saved test_out_final.csv")
```

```python
# assume df (train) exists and meta_preds is your meta predictions for train/test
# compute brand median on train
brand_med = df.groupby('brand_clean')['price'].median().to_dict()

def brand_median_fallback(preds, df_rows, brand_med, floor_ratio=0.15, ceil_ratio=10.0):
    preds2 = preds.copy()
    for i, row in df_rows.iterrows():
        bm = brand_med.get(row['brand_clean'])
        if bm is None: continue
        if preds2[i] < bm*floor_ratio or preds2[i] > bm*ceil_ratio:
            preds2[i] = bm  # or set to (pred+bm)/2 for softer fallback
    return preds2

# Example: apply to your meta test preds (or OOF to measure effect)
meta_oof_fixed = brand_median_fallback(meta_oof.copy(), df.reset_index(drop=True), brand_med)
print("SMAPE after brand-median fallback:", smape(y_raw, meta_oof_fixed))
```

```python
# ============================================================
# Per-Bin Specialized Models (3 macro bins)
# cheap (0–20%), middle (20–80%), premium (80–100%)
# ============================================================

import numpy as np, pandas as pd, lightgbm as lgb, joblib
from scipy.sparse import csr_matrix, hstack

print("Building macro bins from OOF meta predictions...")

# Use your stacked + calibrated OOF predictions (preds_cal_oof)
# If you named it differently, adjust here:
meta_oof = meta_oof_cal.copy()  # <--- IMPORTANT
y_real = df['price'].values

# Compute quantile thresholds
low_thr = np.quantile(meta_oof, 0.20)
high_thr = np.quantile(meta_oof, 0.80)
print("Bin thresholds: low_thr =", low_thr, ", high_thr =", high_thr)

# Assign bins
def assign_macro_bin(pred):
    if pred <= low_thr: return "cheap"
    if pred >= high_thr: return "premium"
    return "middle"

df['macro_bin'] = [assign_macro_bin(p) for p in meta_oof]

df['macro_bin'].value_counts()
```

```python
# ============================================================
# TRAIN CHEAP BIN MODEL
# ============================================================

cheap_idx = df[df['macro_bin']=="cheap"].index
df_ch = df.loc[cheap_idx]

print("Training cheap-bin model on:", len(df_ch), "rows")

X_ch_text = tf.transform(df_ch['catalog_content'])
X_ch_num = csr_matrix(scaler.transform(df_ch[num_features].fillna(-1)))
X_ch = hstack([X_ch_text, X_ch_num])
y_ch = np.log1p(df_ch['price'].values)

dtrain_ch = lgb.Dataset(X_ch, label=y_ch)

params_ch = {
    'objective':'regression',
    'metric':'rmse',
    'learning_rate':0.03,
    'num_leaves':128,
    'feature_fraction':0.8,
    'bagging_fraction':0.8,
    'bagging_freq':5,
    'seed':42
}

cheap_model = lgb.train(
    params_ch, dtrain_ch, num_boost_round=1200,
    valid_sets=[dtrain_ch],
    callbacks=[lgb.early_stopping(stopping_rounds=80)],
)

print("Cheap model best iteration:", cheap_model.best_iteration)
```

```python
# ============================================================
# TRAIN PREMIUM BIN MODEL
# ============================================================

premium_idx = df[df['macro_bin']=="premium"].index
df_pr = df.loc[premium_idx]

print("Training premium-bin model on:", len(df_pr), "rows")

X_pr_text = tf.transform(df_pr['catalog_content'])
X_pr_num = csr_matrix(scaler.transform(df_pr[num_features].fillna(-1)))
X_pr = hstack([X_pr_text, X_pr_num])
y_pr = np.log1p(df_pr['price'].values)

dtrain_pr = lgb.Dataset(X_pr, label=y_pr)

params_pr = {
    'objective':'regression',
    'metric':'rmse',
    'learning_rate':0.03,
    'num_leaves':128,
    'feature_fraction':0.8,
    'bagging_fraction':0.8,
    'bagging_freq':5,
    'seed':42
}

premium_model = lgb.train(
    params_pr, dtrain_pr, num_boost_round=1200,
    valid_sets=[dtrain_pr],
    callbacks=[lgb.early_stopping(stopping_rounds=80)],
)

print("Premium model best iteration:", premium_model.best_iteration)
```

```python
# ============================================================
# OOF Macro Routing (evaluate improvement)
# ============================================================

oof_macro = meta_oof.copy()

print("Replacing cheap & premium with specialized model predictions...")

# CHEAP
X_ch_full_text = tf.transform(df.loc[cheap_idx]['catalog_content'])
X_ch_full_num = csr_matrix(scaler.transform(df.loc[cheap_idx][num_features].fillna(-1)))
X_ch_full = hstack([X_ch_full_text, X_ch_full_num])
pred_ch = np.expm1(cheap_model.predict(X_ch_full))
oof_macro[cheap_idx] = pred_ch

# PREMIUM
X_pr_full_text = tf.transform(df.loc[premium_idx]['catalog_content'])
X_pr_full_num = csr_matrix(scaler.transform(df.loc[premium_idx][num_features].fillna(-1)))
X_pr_full = hstack([X_pr_full_text, X_pr_full_num])
pred_pr = np.expm1(premium_model.predict(X_pr_full))
oof_macro[premium_idx] = pred_pr

# Evaluate
oof_macro_smape = smape(y_real, oof_macro)
print("OOF SMAPE after per-bin specialization:", oof_macro_smape)
```
